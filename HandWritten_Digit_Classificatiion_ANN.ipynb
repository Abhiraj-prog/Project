```python
import pandas as pd
import numpy as np
import tensorflow
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense,Flatten
```

```python
(X_train,y_train),(X_test,y_test)= keras.datasets.mnist.load_data()
```

```python
X_train[0]
```

```python
X_train.shape
```

```python
import matplotlib.pyplot as plt
plt.imshow(X_train[0])
```

```python
# Normalize pixel values from range 0-255 to 0-1 for better model training
X_test=X_test/255
X_train=X_train/255
```

```python
model=Sequential()
# Convert 28x28 image matrix into a single 784-length vector for neural network input
model.add(Flatten(input_shape=(28,28)))
model.add(Dense(128,activation='relu'))
model.add(Dense(32,activation='relu'))
# agar out mein ek se jyada nodes toh softmax use karna hai
model.add(Dense(10,activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(loss='sparse_categorical_crossentropy',optimizer='Adam',metrics=['accuracy'])
```

```python
from IPython.core import history
history=model.fit(X_train,y_train,epochs=25,validation_split=0.2)
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
```

```python
# Predict class probabilities for each test sample
y_prob = model.predict(X_test)

# Select the class with highest probability as final prediction
y_pred = y_prob.argmax(axis=1)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
